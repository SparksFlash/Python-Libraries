### 2D Line plot

image.png

- Bivariate Analysis (On 2 columns of a dataset)
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
#plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
plt.grid()
```

```python
# From a pandas DataFrame
batsman = pd.read_csv('/home/sourav/Python-Libraries/datasets/sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# Plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
# plt.style.use('default')
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'], color='Green')
plt.plot(batsman['index'],batsman['RG Sharma'], color='Red')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Marker snd size
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2, marker='.')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
 # legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='upper left')
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('/home/sourav/Python-Libraries/datasets/batter.csv')
df = df.head(50)
df
```

```python
# size
tips = sns.load_dataset('tips')

# which plots dot is bigger those are with more size
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*20)
```

```python
plt.plot(tips['total_bill'], tips['tip'],'o')
# scatter plot with plot which is not editable
```

```python
# Here s is size of the marker
plt.scatter(x,y,marker='+',color='red',s=50)
```

### Bar chart

image.png

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red', 'blue', 'green', 'yellow', 'orange']

plt.bar(colors, children, color='red')
```

```python
# horizontal bar chart
plt.barh(colors,children,color='black')
```

```python
df = pd.read_csv('/home/sourav/Python-Libraries/datasets/batsman_season_record.csv')
df 
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='lightblue')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='green')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='orange')

plt.xticks(np.arange(df.shape[0]), df['batsman']) # xtricks for x-axis labels

plt.show()
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
```

```python
# Stacked Bar chart
# Here we are stacking the bars on top of each other
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')     # bottom is the previous bar
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','bangla']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('/home/sourav/Python-Libraries/datasets/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'],autopct='%0.1f%%')  # autopct is used to show percentage on pie chart
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
# explode is used to separate the slices of pie chart

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
